# Seaborn Library of Matplotlib
## seaborn is a data visualization library in Matplotlib that provides high level interface for creating interactive and more colorful plots with less code.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
palette=sns.color_palette()
print(palette)
```

```python
sns.palplot(palette)
```

```python
palette_blue=sns.color_palette('Blues',20)
sns.palplot(palette_blue)
```

```python
data_dict={
    'building':['burj-khalifa','merdeka118','Eiffel tower','KLCC'],
    'category':['highest','medium','low','highest'],
    'building_height':[150,120,100,180]
}
bar_colors=['red','green','yellow']
data=pd.DataFrame(data_dict)
sns.barplot(x='building',y='building_height',data=data,hue='category',palette=bar_colors)
plt.show()
```

```python
data_iris=sns.load_dataset('iris')
data_iris
#data_flights=sns.load_dataset('flights')
#data_flights
```

```python
sns.barplot(x='species',y='sepal_width',data=data_iris,hue='species',palette=bar_colors)
plt.title('Iris Species distribution Bar Chart')
plt.xlabel('Species Name')
plt.ylabel('Sepal Width')
plt.grid(axis='y',linestyle='--',color='lightblue',alpha=0.5)
plt.show
```

```python
sns.violinplot(x='species',y='sepal_width',data=data_iris,hue='species',palette=bar_colors)
plt.title('Iris Species distribution Bar Chart')
plt.xlabel('Species Name')
plt.ylabel('Sepal Width')
plt.grid(axis='y',linestyle='--',color='lightblue',alpha=0.5)
```

```python
sns.violinplot(x='species',y='sepal_width',data=data_iris,hue='species',palette='Set2')
plt.title('Iris Species Distribution Bar Chart')
plt.xlabel('Name Species')
plt.ylabel('Sepal Width')
plt.grid(axis='y',linestyle='--',color='lightblue', alpha=0.5)
```

```python
sns.stripplot(x='species',y='sepal_width',data=data_iris,hue='species',palette='Set2')
plt.title('Iris Species Distribution Bar Chart')
plt.xlabel('Name Species')
plt.ylabel('Sepal Width')
plt.grid(axis='y',linestyle='--',color='lightblue', alpha=0.5)
```

```python
data_iris=sns.load_dataset('iris')
sns.pairplot(data=data_iris,hue='species')
plt.show()
```

```python
sns.set_style('darkgrid')
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=data_iris)
plt.title('Darkgrid Theme example using Data Iris')
plt.show()
```

```python
sns.set_style('dark')
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=data_iris)
plt.title('Dark Theme example using Data Iris')
plt.show()
```

```python
#darkgrid,whitegrid,dark,white,ticks
sns.set_style('ticks')
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=data_iris)
plt.title('Ticks Theme example using Data Iris')
plt.xlabel('Sepal Length in cm')
plt.ylabel('Sepal width in cm')
plt.xticks([4,5,6,7,8],fontsize=10,color='magenta')
plt.yticks([1,2,3,4,5,6,7,],fontsize=8,color='purple')
sns.despine()
plt.show()
```
