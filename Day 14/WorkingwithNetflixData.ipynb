```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sns.set_theme(style='whitegrid',palette='pastel')
sns.countplot(data=df,x='type')
plt.show()
```

```python
sns.set_theme(style='whitegrid',palette='deep')
sns.countplot(data=df,x='type')
plt.show()
```

# Use Netflix dataset to visualize the distribution of release years for Movies vs TV Shows using Seaborn's histplot()
## Use different color for Movie Vs TV Show

```python
df = df.dropna(subset=['release_year', 'type'])
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='release_year', hue='type', multiple='stack', palette={'Movie': 'skyblue', 'TV Show': 'salmon'}, bins=30)
plt.title('Distribution of Release Years: Movies vs TV Shows')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
