# Matplotlib.pyplot.subplots
    subplots(): provides a way to plot multiple plots in a single figure
    given number of rows and columns. It returns a tuple (fig,ax) giving a single figure fig with an array of axes ax.

```python
import matplotlib.pyplot as plt
```

```python
sales_data={'Electronics':45000,'Clothing':30000,'Grocery':25000,'Mobile':20000}
category=list(sales_data.keys())
actual_sales=list(sales_data.values())
print(category)
print(actual_sales)
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
x=np.random.random(10)
ax1.plot(x,x/2)
ax2.scatter(np.random.random(20),np.random.random(20))
ax3.bar(category,actual_sales)
ax4.hist(np.random.randn(1000))
```

## Subplots using Indexing

```python
our_lines=np.linspace(0,10,100)
```

```python
product_prices={'Hair Oil':20, 'Face Wash':30, 'Lip balm':10, 'Face Cream':15}
```

```python
our_numbers=np.random.random(10)
```

```python
our_dots=np.random.randn(1000)
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(our_lines,our_lines/2,color='maroon')
ax[0,1].bar((product_prices.keys()),list(product_prices.values()),color=['pink','magenta','hotpink','purple'])
ax[1,0].scatter(np.random.randn(100),np.random.randn(100),color='orange')
ax[1,1].hist(our_dots,color='violet')
plt.tight_layout()
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(our_lines,our_lines/2,color='maroon')
ax[0,1].bar((product_prices.keys()),list(product_prices.values()),color=['pink','magenta','hotpink','purple'])
ax[1,0].scatter(np.random.randn(100),np.random.randn(100),color='orange')
ax[1,1].hist(our_dots,color='violet')

ax[0,0].set_title('Line Plot Example')
ax[0,0].set_xlabel('X Axis')
ax[0,0].set_ylabel('Y Axis')
ax[0,0].grid(True,linestyle='--',alpha=0.7,color='lightblue')

ax[0,1].set_title('Bar Chart Example')
ax[0,1].set_xlabel('Product')
ax[0,1].set_ylabel('Sales')
ax[0,1].grid(True,linestyle='--',alpha=0.7,color='lightblue',axis='y')

ax[1,0].set_title('My Scatter Plot')
ax[1,0].grid(True,linestyle='--',alpha=0.7,color='lightblue')

ax[1,1].set_title('Histogram Example')
ax[1,1].grid(True,linestyle='--',alpha=0.7,color='lightblue',axis='y')

fig.suptitle('Matplot Lib SubPlots example using 2 rows and 2 cols')
plt.tight_layout()
plt.savefig('subplotex.jpeg',dpi=300,bbox_inches='tight')
```
