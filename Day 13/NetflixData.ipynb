```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('netflix_data.csv')
print(df.head())
```

```python
df=df.dropna(subset=['release_year'])
```

```python
df['release_year']=df['release_year'].astype(int)
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['release_year'],bins=100,color='maroon',edgecolor='magenta',alpha=0.8)
plt.title('Netflix Release Title by Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(color='blue',linestyle='--',linewidth=0.7)
plt.show()
```

```python
df=df.dropna(subset=['release_year','type'])
```

```python
df['release_year']=df['release_year'].astype(int)
```

```python
movies=df[df['type']=='Movie']
shows=df[df['type']=='TV Show']
```

```python
plt.figure(figsize=(10,5))
plt.hist([movies['release_year'],shows['release_year']],
         bins=300,
         color=['red','cyan'],
         edgecolor='purple',
         alpha=0.8,
         label=['Movies','TV Shows']
)
plt.title('Netflix Release by Year and Type')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(color='blue',linestyle='--',linewidth=0.7)
plt.tight_layout()
plt.legend(title='Movies/shows')
plt.savefig('netflix_showmovies.jpg',dpi=300,bbox_inches='tight')
plt.show()
```

# Use netflix_data.csv dataset
### print how many Movies and TV Shows are available on Netflix
### display movies and TV Shows proportion in a pie plot
### Save file as pdf

```python
df=df.dropna(subset=['type'])
```

```python
type_counts = df['type'].value_counts()
print(type_counts)
```

```python
colors = plt.cm.rainbow([0.3, 0.6])
```

```python
plt.figure(figsize=(6,6))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Distribution of Movies and TV Shows on Netflix')
plt.savefig('netflix_distribution.pdf')
plt.show()
```
