```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
temp=np.random.randint(20,100,size=50)
plt.scatter(x,y,c=temp,cmap='plasma',s=120,edgecolor='black')
plt.colorbar(label='Temperature')
plt.title('Scatter plot Colored by Temperature')
plt.xlabel('x-Random')
plt.ylabel('y-Random')
plt.show()
```

```python
height_cm=[150,155,160,165,170,175,180,185]
weight_kg=[50,53,56,59,62,66,68,72]
age_year=[20,22,25,28,30,35,38,40]
```

```python
plt.figure(figsize=(8,5))
scatter=plt.scatter(height_cm,weight_kg,s=120,c=age_year,cmap='rainbow',marker='*',alpha=0.9)
plt.colorbar(scatter,label='Age(years)')
plt.title('Scatter plot: Height vs Weight(Colored by Age)')
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.xlim(140,190)
plt.ylim(40,90)
plt.grid(color='pink',alpha=0.5,linestyle='--',linewidth=0.5)
plt.xticks(rotation=45, fontsize=12, color='orange')
plt.yticks([50,55,60,65,70,75,80],['50kg','55kg','60kg','65kg','70kg','75kg','80kg'],fontsize=12, color='hotpink')
plt.show()
```

# Pie Plot Example
    plt.pie(x,labels='',colors='',autopct='',startangle='',shadow=,explode=,radius=

```python
depts=['HR','Web-Development','App-Development','Agile-Scrum','Testing']
expenses=[12000,25000,18000,15000,20000]
```

```python
plt.figure(figsize=(7,7))
```

```python
plt.pie(expenses,labels=depts,autopct='%1.2f%%',startangle=140,shadow=True,explode=(0,0.05,0,0,0))
plt.title('Department Expense Report')
plt.show()
```

```python
plt.pie(expenses,labels=depts,colors=plt.cm.rainbow([0.1,0.3,0.5,0.7,0.9]),autopct='%1.2f%%',startangle=140,shadow=True,explode=(0,0.05,0,0,0))
plt.title('Department Expense Report')
plt.show()
```

```python
plt.pie(expenses,labels=depts,colors=plt.cm.magma([0.1,0.3,0.5,0.7,0.9]),autopct='%1.2f%%',startangle=140,shadow=True,explode=(0,0.05,0,0,0))
plt.title('Department Expense Report')
plt.show()
```

```python
plt.pie(expenses,labels=depts,colors=plt.cm.inferno([0.1,0.3,0.5,0.7,0.9]),autopct='%1.2f%%',startangle=140,shadow=True,explode=(0,0.05,0,0,0))
plt.title('Department Expense Report')
plt.show()
```

# create pie plot
    sales_data={'Mobile':45000,'Clothing':30000,'Groceries':35000,'Home Decor':20000}
    show percentage with two decimal
    use colormap to give different color of pie

```python
sales_data = {'Mobile': 45000, 'Clothing': 30000, 'Groceries': 35000, 'Home Decor': 20000}
labels = list(sales_data.keys())
sizes = list(sales_data.values())
```

```python
plt.figure(figsize=(7,7))
colors = plt.cm.rainbow([0.1, 0.3, 0.5, 0.7])
```

```python
plt.figure(figsize=(7,7))
plt.pie(sizes, labels=labels, autopct='%.2f%%', colors=colors, startangle=140, textprops={'color':'white','fontsize':10,'fontweight':'bold'}),
plt.title('Sales Distribution by Category')
#plt.savefig('filenamewithextension',dpi=value,bbox_inches='')
plt.show()
```

```python
marks=[45,60,80,55,62,90,89,50,56,60,80,55,62,90,89,58,90,56,67,83]
plt.figure(figsize=(8,5))
plt.hist(marks,bins=5,color='skyblue',edgecolor='darkblue',alpha=0.8)
plt.title('Histogram of Student Marks')
plt.xlabel('Marks Range')
plt.ylabel('Number of Students')
plt.grid(color='blue',linestyle='--',linewidth=0.7)
plt.show()
```
