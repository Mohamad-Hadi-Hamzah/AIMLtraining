```python
import pandas as pd
```

```python
# Sample dataset
data = {
    'Name':['Jennie','Lalisa','Rose',None,'Jisoo'],
    'Age':[25,None,30,22,None],
    'City':['KL','Penang',None,'Johor','KL']
}
```

```python
dt=pd.DataFrame(data)
print('Original Dataframe:')
dt
```

# Check missing values row by row

```python
for index, row in dt.iterrows():
    print(f'Row {index}: Missing = {row.isnull().any()}, Details = {row.isnull().to_dict()}')
```

# Show only rows with missing data

```python
missing_rows = dt[dt.isnull().any(axis=1)]
print('Rows with missing data: ')
missing_rows
```

# Simulate rows that would be dropped using dropna()

```python
to_drop = dt[dt.isnull().any(axis=1)]
print('These rows would be dropped using dropna(): ')
to_drop
```

```python
dt_cleaned = dt.dropna()
print('After dropna(): ')
dt_cleaned
```

```python
print('Before fillna(): ')
dt
```

```python
dt_filled = dt.fillna({
    'Name':'Unknown',
    'Age':dt['Age'].mean(),
    'City':'Not Available'
})
```

```python
print('After fillna(): ')
dt_filled
```

```python
dt_sfill = dt.fillna({
    'Name':'Unknown',
    'Age':round(dt['Age'].mean(),2),
    'City':'Not Available'
})
    
```

```python
print('After fillna(): Age up to two decimal: ')
dt_sfill
```
