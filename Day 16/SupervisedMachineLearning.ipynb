# Supervised Machine Learning
    Linear Regression

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
data = pd.read_csv('sales_data.csv')
data.head()
```

```python
X = data[['Advertising_Spend']]
y = data['Sales_Revenue']
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
new_spend = float(input('Enter New Spend to Predict Sales Avenue'))
predicted_revenue = model.predict([[new_spend]])
print(f'Predicted sales revenue for RM{new_spend} ad spend: RM{predicted_revenue[0]:.2f}')
```

# Linear Regression Excercise

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
dt = pd.read_csv('house_price.csv')
dt = dt[dt['Price'] != '?'] #removed the row with empty price
dt
```

```python
A = dt[['Size_sqft']]
b = dt['Price']
```

```python
model = LinearRegression()
model.fit(A, b)
```

```python
pr_price=model.predict([[1700]])
print(pr_price)
```
