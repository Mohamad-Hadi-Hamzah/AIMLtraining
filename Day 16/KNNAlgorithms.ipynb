# KNN Algorithm is Supervised Machine Learning
# KNeighbourClassifier of sklearn library for this one

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = np.array([
    [1, 3500],
    [2, 3200],
    [3, 3000],
    [4, 2500],
    [5, 2300],
    [6, 2000],
    [7, 2200],
    [8, 3700],
])
y = ['Unfit','Unfit','Unfit','Fit','Fit','Fit','Fit','Unfit']
```

```python
model=KNeighborsClassifier(n_neighbors=3)
```

```python
model.fit(X,y)
```

```python
new_person=[[4.5,2600]]
pred=model.predict(new_person)
print('Predicted result: ',pred[0])
```

```python
exercise_hours=float(input('Enter Number of Exercise Hours: '))
cal_intake=int(input('Enter Number of Calories Consumed: '))
pred_fu=model.predict([[exercise_hours,cal_intake]])
print(f'Exercise Hours {exercise_hours}, Calories Intake {cal_intake}, You are likely {pred_fu[0]}')
```

# Downlaod student_grade.csv from GitHub.com
    Predict a student's grade level (A, B, or C)
    based on hours studied and assignments completed
    using the K-Nearest Neighbors algorithm

```python
df=pd.read_csv('student_grade.csv')
df
```

```python
effort=df[['Hours_Studied','Assignments_Completed']]
grade=df['Grade']
```

```python
model=KNeighborsClassifier(n_neighbors=3)
```

```python
model.fit(effort,grade)
```

```python
hours=float(input('Enter number of hours studied: '))
assignments=float(input('Enter number of assignment completed: '))

new_grade=pd.DataFrame([[hours,assignments]],columns=['Hours_Studied','Assignments_Completed'])
pred_grade=model.predict(new_grade)
print(f'{hours} hours studied with {assignments} assignments completed, is likely to get {pred_grade[0]} grade')
```
