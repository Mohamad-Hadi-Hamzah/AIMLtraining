```python
import pandas as pd
```

```python
df=pd.read_csv('patients.csv')
print(df.head())
```

```python
age_counts=df['age'].value_counts().sort_index()
plt.figure(figsize=(8,5))
plt.bar(age_counts.index,age_counts.values,color='skyblue',width=0.5,edgecolor='orange')
plt.title('Number of Patients by Age')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.show()
```

# Create a Bar Plot - Average age by patients by Arrival date

```python
df['arrival_date']=pd.to_datetime(df['arrival_date'])                                
avg_age_bydate=df.groupby('arrival_date')['age'].mean()
```

```python
plt.figure(figsize=(10,5))
plt.bar(avg_age_bydate.index.astype(str),avg_age_bydate.values,color='cyan',width=0.5,edgecolor='magenta')
plt.title('Average Patient Age by Arrival Date')
plt.xlabel('Arrival Date')
plt.ylabel('Average Age')
plt.show()
```

# Scatter Plots
    In scatter plot we represent data in point, where each point represent a data observation, we can check how the values corelate.

```python
Product_sales={'Laptop':750, 'Smart Phone':650, 'Smart Watch':400, 'Tablets':500, 'AC':400}
products = list(Product_sales.keys())
sales = list(Product_sales.values())
plt.scatter(products,sales)
plt.title('Basic product and sales representation')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.grid()
plt.show()
```

```python
players=['Jennie','Lalisa','Jisoo','Rose','IU']
match1_score=[45,22,77,89,20]
match2_score=[10,60,66,33,90]
match3_score=[15,68,88,99,88]
plt.figure(figsize=(8,5))
plt.scatter(players,match1_score,color='pink',label='Match One',s=100)
plt.scatter(players,match2_score,color='purple',label='Match Two',s=100, marker='x')
plt.scatter(players,match3_score,color='cyan',label='Match Three',s=100, marker='*')
plt.title('Comparison of Players in Two Matches')
plt.xlabel('Players')
plt.ylabel('Score')
plt.legend(loc='lower center')
plt.grid(True,linestyle='--',alpha=0.75)
plt.show()
```

## Create a bar chart using following data
### with following libraries
    python_lib=['Numpy','Pandas','Matplotlib','Scikit-learn']
### liked by users as follows
    liked_by=[5000,7000,9000,4000]
## Create a vertical bar graph where each bar represents the value of python_lib respective liked by users
    every bar must have a different color

```python
python_lib=['Numpy','Pandas','Matplotlib','Scikit-learn']
liked_by=[5000,7000,9000,4000]
rainbow=['pink','magenta','purple','orange']
```

```python
plt.figure(figsize=(8,5))
plt.bar(python_lib,liked_by,color=rainbow)
plt.title('Python Libraries Popularity')
plt.xlabel('Library')
plt.ylabel('Liked by Users')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.barh(python_lib,liked_by,color=rainbow)
plt.title('Python Libraries Popularity')
plt.xlabel('Library')
plt.ylabel('Liked by Users')
plt.show()
```
