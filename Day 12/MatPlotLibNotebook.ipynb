    ##Figure :The complete window to create plot
    ##Axes   :The area where data is plotted
    ##Axis   :x line and y line to define scale and ticks
    ##Plot   :The graph of data

# First example to create line chart from data
    Students.xlsx for fetching the data
    Generate line chart based on student name and student marks

```python
import pandas as pd
```

```python
df=pd.read_excel('/Users/marinacherani/Downloads/AIML/AIMLtraining/Day 10/students.xlsx')
```

```python
names=df['Name']
marks=df['Marks']
```

```python
plt.plot(names,marks,marker='*',color='pink')
plt.title('Student Marks Line Chart')
plt.xlabel('Student Name')
plt.ylabel('Student Marks')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.plot(names,marks,marker='h',color='cyan')
plt.title('Student Progress Chart')
plt.xlabel('Student Name')
plt.ylabel('Student Marks')
plt.grid(True)
plt.show()
```

### Linestyle options
    : , -. , -- 

```python
print('Dashed Line')
plt.plot(names,marks,marker='v',color='orange',linestyle='--')
plt.show()
```
